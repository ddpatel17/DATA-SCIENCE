```python
import pandas as pd
```

```python
s1 = pd.read_csv("../../Data/marine economy.csv")
s1.head()
```

```python
result = pd.DataFrame(s1)
result
```

```python
s1['data_value'].agg(['mean','sum','count'])
```

```python
s1.agg({'data_value':['mean','sum'],'year':['count']})
```

```python
data = { 'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'Salary': [50000, 60000, 75000] }
df = pd.DataFrame(data)
print(df)
agg_results = df.agg({ 'Salary': ['mean', 'sum'], 'Age': ['mean', 'count'] })
agg_flat = {
    'Average_Salary': agg_results.loc['mean', 'Salary'], 'Total_Salary': agg_results.loc['sum', 'Salary'],
    'Average_Age': agg_results.loc['mean', 'Age'], 'Employee_Count': agg_results.loc['count', 'Age'] }
agg_df = pd.DataFrame([agg_flat])
print(agg_df)
```

```python
# aggregate() Applies aggregation functions to grouped data.
# df.agg(func)
```

```python
group = s1.groupby('year').agg({'data_value':'mean', 'units':'count'}).rename(columns={'data_value':'averaage_data_value','units':'count_units'})
print(group)
```

```python
s1_melted = s1.melt(id_vars='year',var_name='attributes',value_name='value')
print(s1_melted)
```

```python
s2_melted = s1.melt(id_vars='year',var_name='attributes',value_name='value')
print(s2_melted)
```

```python
s2_melted = s2_melted.drop_duplicates(subset=['year','attributes']) # without drop_duplicate get error 
s1_pivot = s2_melted.pivot(index='year',columns='attributes',values='value')
print("\n pivoted DataFrame")
print(s1_pivot)
```

```python
s1.pivot_table(index='year',values='data_value',aggfunc='mean')
```

```python
s1['data_value'].transform(lambda x: (x-x.min())/(x.max()-x.min()))
```

```python
data = {
'Name': ['Alice', 'Bob', 'Charlie'],
'Age': [25, 30, 35],
'Salary': [50000, 60000, 75000]}
df = pd.DataFrame(data)

df['Normalized_Salary'] = df['Salary'].transform(lambda x: x / df['Salary'].max())
#print("\nDataFrame with Normalized Salary:")
print(df)
```

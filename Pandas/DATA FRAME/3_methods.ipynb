```python
import pandas as pd
```

```python
s1 = pd.read_csv("../../Data/marine economy.csv")
s1.head()
```

```python
result = pd.DataFrame(s1)
result
```

```python
d = s1.loc[s1['data_value']>30,['year','category','flag','variable','data_value']]
d
```

```python
d.iloc[0:3,1:3] 
```

```python
d.at[7,'year'] # row label and columns label
```

```python
d.iat[0,3] # integer row position integer column position
```

```python
s1.query('data_value > 30 and year == 2014')
```

```python
s1.filter(items=['category'],axis=0) # for axis is 0 only header line (column) are print.
```

```python
s1.filter(items=['0'],axis=1)
```

```python
s1.filter(items=['data_value','year'],axis=1) # for axis is 1 items are print.
```

```python
s1.select_dtypes(include=float,exclude=int)
```

```python
s1.select_dtypes(include=float,exclude=None)
```

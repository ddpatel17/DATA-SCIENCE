```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit = sns.load_dataset('titanic')
tit.head()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.swarmplot(data=tit,x='class',y='age')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.swarmplot(data=tit,x='class',y='age',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age',hue='sex',jitter=0.4)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.swarmplot(data=tit,x='class',y='age',hue='sex') # jitter not supported in swrmplot.
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age',hue='sex',jitter=0.3,dodge=True,size=3)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.swarmplot(data=tit,x='class',y='age',hue='sex',dodge=True,size=3)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='age',y='class',hue='sex',jitter=0.3,dodge=True,size=3)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.swarmplot(data=tit,x='age',y='class',hue='sex',dodge=True,size=3)
plt.show()
```

```python
plt.figure(figsize=(9,4),dpi=90)
sns.swarmplot(data=tit,x='age',y='class',hue='sex',dodge=True,size=3)
plt.show()
```

```python
sns.swarmplot(data=tit, x='deck', y="fare", hue="sex", dodge=True,size=3)
plt.title("Fare by Embarkation Port and Sex")
plt.show()
```

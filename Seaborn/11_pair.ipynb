```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pairplot(data=tips)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pairplot(data=tips,hue='smoker')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pairplot(data=tips,hue='sex',diag_kind='hist')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pairplot(data=tips,hue='smoker',diag_kind='kde')
plt.show()
```

```python
tit=sns.load_dataset('titanic')
tit.head()
```

```python
plt.figure(figsize=(6,4),dpi=60)
titanic_num = tit[['age', 'fare', 'sibsp', 'parch', 'survived','who']].dropna()
sns.pairplot(titanic_num,hue='who')
plt.suptitle('Pair Plot of Titanic Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
titanic_num = tit[['age', 'fare', 'sibsp', 'parch', 'survived','who']].dropna()
sns.pairplot(titanic_num,hue='who')
plt.suptitle('Pair Plot of Titanic Numerical Features',y=1.05)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
titanic_num = tit[['age', 'fare', 'sibsp', 'parch', 'survived','who']].dropna()
sns.pairplot(titanic_num,hue='who',palette='Set3') # palette colour is Set1(dark), Set2(medium), Set3(light)
plt.suptitle('Pair Plot of Titanic Numerical Features',y=1.05)
plt.show()
```

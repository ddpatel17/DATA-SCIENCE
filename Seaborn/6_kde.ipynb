```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit = sns.load_dataset('titanic')
tit.head()
```

```python
plt.figure(figsize=(6,4),dpi=(60))
sns.kdeplot(x='fare',data=tit,fill=True,bw_adjust=0.7)
plt.show()
```

```python
age1=tit.dropna(subset=['age'])
plt.figure(figsize=(6,3),dpi=(60))
sns.kdeplot(x='age',data=age1,fill=True)
plt.show()
```

```python
age1=tit.dropna(subset=['age'])
plt.figure(figsize=(6,4),dpi=(60))
sns.kdeplot(x='age',data=age1,fill=True,hue='sex')
plt.show()
plt.figure(figsize=(6,4),dpi=(60))
sns.kdeplot(x='age',data=age1,fill=True,hue='sex',common_norm=False,alpha=0.7) # alpha means transparency of kde.
plt.show()
```

```python
age1 = tit.dropna(subset=['age'])
plt.figure(figsize=(6,4),dpi=(60))
sns.kdeplot(x='age',data=age1,fill=True,hue='class')
plt.show()
```

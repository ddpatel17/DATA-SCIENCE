```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit = sns.load_dataset('titanic')
tit.columns
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age',hue='sex',jitter=0.4)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='class',y='age',hue='sex',jitter=0.3,dodge=True,size=3)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.stripplot(data=tit,x='age',y='class',hue='sex',jitter=0.3,dodge=True,size=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit = sns.load_dataset('titanic')
tit.head()
```

```python
tit = sns.load_dataset('titanic')
tit.columns
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.violinplot(x='who',y='fare',data=tit)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.violinplot(x='who',y='fare',data=tit,inner='points')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.violinplot(x='who',y='fare',data=tit,hue='survived',inner='points')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.violinplot(x='who',y='fare',data=tit,hue='survived',split=True)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.violinplot(x='who',y='age',data=tit,hue='survived',split=True)
plt.show()
```

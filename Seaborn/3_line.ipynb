```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit=sns.load_dataset('titanic')
tit.head()
```

```python
tit1=tit.dropna(subset=['age','survived'])
```

```python
plt.figure(figsize=(6,4),dpi=(60))
sns.lineplot(x='age',y='survived',data=tit1)
plt.title('average survived by age')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=(60))
sns.lineplot(x='age',y='survived',data=tit1,hue='who')
plt.title('average survived by age')
plt.show()
```

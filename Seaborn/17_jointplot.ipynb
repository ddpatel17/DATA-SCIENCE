```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(6,3),dpi=80)
sns.jointplot(data=tips,x='tip',y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(6,3),dpi=80)
sns.jointplot(data=tips,x='tip',y='total_bill',hue='day')
plt.show()
```

```python
plt.figure(figsize=(6,3),dpi=80)
sns.jointplot(data=tips,x='tip',y='total_bill',hue='sex',kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(6,3),dpi=80)
sns.jointplot(data=tips,x='tip',y='total_bill',kind='kde',hue='time')
plt.show()
```

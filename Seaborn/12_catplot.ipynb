```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit = sns.load_dataset('titanic')
tit.columns
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.catplot(x='class',y='age',data=tit)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.catplot(x='class',y='age',data=tit,kind='bar',hue='who')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.catplot(x='class',y='age',data=tit,kind='bar',hue='who')
plt.ylim(0,70)
plt.show()
```

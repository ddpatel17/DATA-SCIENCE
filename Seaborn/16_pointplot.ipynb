```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit = sns.load_dataset('titanic')
tit.head()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pointplot(data=tit,x='class',y='age')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pointplot(data=tit,x='class',y='age',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pointplot(data=tit,x='class',y='age',hue='sex',dodge=0.3)
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pointplot(data=tit,x='class',y='age',hue='sex',dodge=0.3,estimator='std')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=60)
sns.pointplot(data=tit,x='age',y='class',hue='sex',dodge=0.7,estimator='std')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tit = sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(6,4),dpi=(60))
sns.histplot(x='age',data=tit,bins=16)
plt.title("distribution by age")
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=(60))
sns.histplot(x='age',data=tit,bins=16,kde=True,color='red')
plt.title("distribution by age")
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=(60))
sns.histplot(x=tit['age'].dropna(),data=tit,bins=16,kde=True,color='green')
plt.title("distribution by age")
plt.show()
```
